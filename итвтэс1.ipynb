## Задание 2
Составить программу для таблицы аргументов и значений функции Z на отрезке [х0, xk] с шагом Δх, записать таблицу в файл. На Python можно использовать numpy или pandas.

```python
x0 = 0.5
xk = 4
xdet = 0.5
a = np.arange(x0, xk + xdet, xdet)
def Z(x):
    return (x+1)**3 - 4*x
y = [Z(x) for x in a]
print(*y)
np.savetxt('data.txt', np.column_stack((a, y)), header='x f(x)', fmt='%.1f')
print(np.loadtxt('data.txt'))
```

Выполните индивидуальное задание из раздела «Чтение и запись структурированных данных» (ниже), для этого напишите необходимые m-функции для чтения числовых данных из файла, их преобразования и записи данных из переменной в файл в виде матрицы.

### Задание для варианта 10:



```python
N = int(input("Введите размер массива: "))
a = [[np.random.randint(0, 10) for _ in range(N)] for _ in range(N)]
for i in range(N):
    for j in range(N):
        if i == j:
            a[i][j] = 0
print(*a, sep="\n")
print(np.savetxt('TURNIR.DAT', a, fmt='%d'))
```

```python
a = np.loadtxt('TURNIR.DAT', dtype=int)

count = sum(a[i][j] for i in range(N) for j in range(N))
max_value = max(sum(a[i]) for i in range(N))
min_value = min(sum(a[i]) for i in range(N))
diff = [sum(a[i]) - sum(a[j][i] for j in range(N)) for i in range(N)]
wins = [sum(1 for j in range(N) if a[i][j] > a[j][i]) for i in range(N)]
losses = [sum(1 for j in range(N) if a[i][j] < a[j][i]) for i in range(N)]
draws = [N - 1 - wins[i] - losses[i] for i in range(N)]
scores = [wins[i]*3 + draws[i] for i in range(N)]

print("Количество забитых голов: ",count)
print("Максимальное количество забитых голов: ", max_value)
print("Минимальное количество забитых голов: ", min_value)
print("Разница между забитыми и пропущенными голами для каждой команды: ", *diff)
print("Количество выигранных матчей для каждой команды:                 ", wins)
print("Количество проигранных матчей для каждой команды:                ", losses)
print("Количество ничьих для каждой команды:                            ", draws)
print("Очки для каждой команды:                                         ", scores)
```
