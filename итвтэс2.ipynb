```python
# Функция с несколькими локальными минимумами
def func(x):
    return np.sin(x[0]) + 0.1 * x[0]**2

# Плохое начальное приближение → локальный минимум 
res1 = minimize(func, x0=[5], method='BFGS')
print(f"Локальный минимум: x = {res1.x[0]:.3f}, f(x) = {res1.fun:.3f}")

# Хорошее начальное приближение → глобальный минимум
res2 = minimize(func, x0=[-1.5], method='BFGS')
print(f"Глобальный минимум: x = {res2.x[0]:.3f}, f(x) = {res2.fun:.3f}")

# Не нужно начальное приближение — ищет глобально
bounds = [(-100, 100)]

res = shgo(func, bounds=bounds, iters=3)
print(f"Глобальный минимум (shgo): x = {res.x[0]:.3f}, f(x) = {res.fun:.3f}")
print(f"Количество вызовов функции: {res.nfev}")
```

# задание 3

Решите задачу многомерной минимизации целевой функции (табл. 4.1) от нескольких аргументов,

Постройте график функции.\
Минимизируйте функцию (табл. 4.1), используя встроенную функцию (в Matlab fminsearch , а в Python scipy.optimize.minimize и scipy.optimize.shgo)



```python
def y(x1, x2):
    return 1.5*x1+ x2**3 + x1**4

x0 = [-5, 7]
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x1, x2)
Z = y(X, Y)

plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=None, alpha=0.95)
ax.set_title('График функции y(x1, x2)')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

res = minimize(lambda x: y(x[0], x[1]), x0, method='BFGS')
print(f"Минимум функции: x1 = {res.x[0]:.3f}, x2 = {res.x[1]:.3f}, y = {res.fun:.3f}")

shgo_res = shgo(lambda x: y(x[0], x[1]), bounds=[(-100, 100), (-100, 100)], iters=3)
print(f"Глобальный минимум (shgo): x1 = {shgo_res.x[0]:.3f}, x2 = {shgo_res.x[1]:.3f}, y = {shgo_res.fun:.3f}")
```

б) исходный ("классический") метод. Оцените как изменилась скорость сходимости и точность решения после модификации:

Вариант  №2. Модификация метода половинного деления. После локализации корня np.abs(b-a) < 10*eps запускается метод полного сканирования с шагом h.\


```python
def f(x):
    return -x/np.exp(x)

x = np.linspace(0,3, 100)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = -x/exp(x)', color='blue')
plt.title('График функции f(x)')

res = minimize(f, x0=[1.5], method='BFGS')
print(f"Минимум функции: x = {res.x[0]:.3f}, f(x) = {res.fun:.3f}")

def full_scan(f, a, b, h):
    start_time = time()
    x_vals = np.arange(a, b + h/2, h)  # +h/2 для включения правой границы
    min_val = np.inf
    for i in range(len(x_vals)):
        val = f(x_vals[i])
        if val < min_val:
            min_val = val
            idx = i
    elapsed = time() - start_time
    print(f"\n   Полный перебор (h={h:.1e}):")
    print(f"   x* = {x_vals[idx]:.3f}, f(x*) = {min_val:.3f}")
    print(f"   Время выполнения: {elapsed:4f} секунд")

full_scan(f, 0, 3, 0.001)

print("Вариант  №2. Модификация метода половинного деления. После локализации корня np.abs(b-a) < 10*eps запускается метод полного сканирования с шагом h.")

def df(x): # производная 
    return np.exp(-x) * (x - 1)

def modified_bisection_scan(df, f, a, b, h):
    start_time = time()
    eps = np.finfo(float).eps
    fa, fb = df(a), df(b)
    
    if fa * fb > 0:
        raise ValueError("Производная не меняет знак на интервале")
    
    # Бисекция
    iter_bisect = 0
    while np.abs(b - a) >= 10 * eps:
        c = (a + b) / 2
        fc = df(c)
        if np.abs(fc) < 1e-14:
            a = b = c
            break
        if fa * fc < 0:
            b, fb = c, fc
        else:
            a, fa = c, fc
        iter_bisect += 1
    
    # Сканирование на маленьком интервале [a, b]
    best_x, min_val = a, f(a)
    steps = max(1, int(np.ceil((b - a) / h)))
    
    for i in range(steps + 1):
        x = min(a + i * h, b)
        val = f(x)
        if val < min_val:
            min_val, best_x = val, x
    
    elapsed = time() - start_time

    print(f"   Модифицированная бисекция (Вариант №2):")
    print(f"   Бисекция: {iter_bisect} итераций, интервал [{a:.2e}, {b:.2e}]")
    print(f"   Сканирование: {steps+1} шагов, шаг h={h:.1e}")
    print(f"   x* = {best_x:.3f}, f(x*) = {min_val:.3f}")
    print(f"   Время выполнения: {elapsed:4f} секунд")
    return best_x, min_val

bisec_res = modified_bisection_scan(df, f, 0, 3, 0.1)
```

# Задание 5

Решите задачу минимизации линейной целевой функции от нескольких аргументов при линейных ограничениях в Python, используя встроенную функцию linprog.



```python
from scipy.optimize import linprog
c = [-70, -50, -100, -80] 

# Столбцы: P1, P2, P3, P4
A_ub = [
    [10,  8, 13,  5],  # S1
    [ 9, 21,  6, 12],  # S2
    [ 8,  9,  7, 11]   # S3
]

b_ub = [420000, 480000, 100000]

x_bounds = (0, None)
bounds = [x_bounds, x_bounds, x_bounds, x_bounds]

res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

if res.success:
    plan = res.x
    print("\n Оптимальный план выпуска продукции:")
    products = ['P1', 'P2', 'P3', 'P4']
    for i, p in enumerate(products):
        print(f"   {p}: {plan[i]:2f} ед.")
    
    max_revenue = -res.fun  
    print(f"\n Максимальная суммарная стоимость: {max_revenue:.2f} руб.")
    
    print("\n Использование ресурсов:")
    used_resources = np.dot(A_ub, plan)
    for i, res_name in enumerate(['S1', 'S2', 'S3']):
        used = used_resources[i]
        total = b_ub[i]
        leftover = total - used
        print(f"   {res_name}: Использовано {used:.2f} из {total} (Остаток: {leftover:.2f})")
        
    print(f"\n Статус решения: {res.message}")
else:
    print("\n Ошибка оптимизации:", res.message)
```
